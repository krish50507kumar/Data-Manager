```python
import pandas as pd
import numpy as np

np.random.seed(42)

# Create base dataframe
df = pd.DataFrame({
    "id": np.arange(1, 51),
    "name": np.random.choice(
        ["Alice", "Bob", "Charlie", "David", "Eva", None],
        50
    ),
    "age": np.random.choice(
        [22, 25, 28, 31, 35, np.nan],
        50
    ),
    "salary": np.random.choice(
        [45000, 55000, 65000, 75000, np.nan],
        50
    ),
    "department": np.random.choice(
        ["HR", "IT", "Finance", "Sales", None],
        50
    ),
    "city": np.random.choice(
        ["Delhi", "Mumbai", "Pune", "Bangalore", None],
        50
    ),
    "rating": np.random.choice(
        [1, 2, 3, 4, 5, np.nan],
        50
    ),
    "experience": np.random.choice(
        [1, 3, 5, 7, 10, np.nan],
        50
    ),
    "status": np.random.choice(
        ["Active", "Inactive", None],
        50
    ),
    "score": np.random.choice(
        [55.5, 67.8, 72.1, 88.4, 91.2, np.nan],
        50
    )
})

# Introduce duplicate rows
df.iloc[10] = df.iloc[0]
df.iloc[20] = df.iloc[5]
df.iloc[30] = df.iloc[10]

print(df)

print("\nShape:", df.shape)
print("\nDuplicate Rows:")
print(df[df.duplicated()])

print("\nNull Values Count:")
print(df.isnull().sum())
```
